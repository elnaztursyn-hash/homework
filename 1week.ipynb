```python
#1
def analyze_text(text):
    glasni="eyuiao"
    text=text.lower()

    clean=""
    for i in range(len(text)):
        if text[i].isalpha() or text[i]==" ":
            clean+=text[i]

    find_glasni=""
    for i in range(len(clean)):
        c=clean[i]
        if c in glasni and c not in find_glasni:
            find_glasni+=c
    con=len(find_glasni)

    words=clean.split()
    result=[]
    for i in range(len(words)):
        w=words[i]
        if len(w)>=5:
            if w[0]==w[-1]:
                if w not in result:
                    result.append(w)
    all=""
    for i in range(len(result)):
        all+=result[i]
        if i != len(result)-1:
            all+=" "
    return (con,all)

text="Aseta Eliyae123456 Matzh Azamata!"
analyze_text(text)


```
