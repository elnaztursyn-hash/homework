```python
#2
#1
def even_numbers(n):
    for i in range(2,n+1,2):
       if i%4==0:
           yield "4 kratno"
       else:
           yield i
for x in even_numbers(10):
    print(x)
```

```python
#2.2
def filter_words(words):
    for x in words:
        if len(x)>=4:
            if "a" in x.lower():
                yield "ca"
            else:
                yield x
nnn=["cat","car","mouse","dogamaster"]
for m in filter_words(nnn):
    print(m)
```

```python
#2.3
def infinite_numbers():
    n=1
    while True:
        if n%3==0 and n%4==0:
            pass
        if n%3==0 and n%5==0:
            yield "FizzBuzz"
        elif n%3==0:
            yield "Fizz"
        elif n%5==0:
            yield "Buzz"
        else:
            yield n
        n+=1
g=infinite_numbers()
for _ in range(15):
    print(next(g))
```

```python
#2.4
def squares(n):
    for i in range(1,n+1):
        a=i**2
        if a%2==0:
            yield "even square"
        else:
            yield a
for x in squares(5):
    print(x)
```

```python
#4.1
def prostoi(x):
    if x<2:
        return False
    for i in range(2,int(x**0.5)+1):
        if x%i==0:
            return False
    return True
def special_numbers(n):
    for i in range(1,n+1):
        if i%3==0 and i%5==0:
            yield "FizzBuzz"
        elif i%3==0:
            yield "Fizz"
        elif i%5==0:
            yield "Buzz"
        elif prostoi(i):
            yield "Prostoi"
        else:
            yield i
for v in special_numbers(15):
    print(v)
```

```python
#4.3
nums= [5, -2, 8, 0, -7, 3]
def process_numbers(numbers):
    filtered=filter(lambda x: x>=0,numbers)
    trans=map(lambda x: x/2 if x%2==0 else x*3+1,filtered)
    for res in trans:
        yield res
for x in process_numbers(nums):
    print(x)
```

```python
#4.5
matrix = [
[1, 2, 3],
[4, 5, 6],
[7, 8, 9]
]
def flatten(matrix):
    for i in matrix:
        for g in i:
            if g%2==0 and g%3==0:
                yield "kratno 6"
            elif g%2==0:
                yield "even"
            elif g%3==0:
                yield "kratno 3"
            else:
                yield g
for x in flatten(matrix):
    print(x)
```
