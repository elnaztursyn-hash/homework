```python
import csv


data = [
    ["name", "department", "salary"],
    ["Ali", "IT", "500000"],
    ["Dana", "HR", "300000"],
    ["Arman", "IT", "600000"],
    ["Aruzhan", "Marketing", "400000"],
    ["Dias", "IT", "450000"]
]

with open('employees.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(data)



employees = []
dept_salaries = {}

with open('employees.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:

        row['salary'] = int(row['salary'])
        employees.append(row)


        dept = row['department']
        if dept not in dept_salaries:
            dept_salaries[dept] = []
        dept_salaries[dept].append(row['salary'])


total_salary = sum(emp['salary'] for emp in employees)
avg_salary = total_salary / len(employees)


dept_avg = {}
for dept, salaries in dept_salaries.items():
    dept_avg[dept] = sum(salaries) / len(salaries)


top_employee = max(employees, key=lambda x: x['salary'])
best_paid_dept = max(dept_avg, key=dept_avg.get)


high_earners = [emp for emp in employees if emp['salary'] > avg_salary]


with open('high_salary.csv', 'w', newline='', encoding='utf-8') as f:
    fieldnames = ['name', 'department', 'salary']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(high_earners)


print(f"Общая средняя зарплата: {avg_salary:.2f}")
print(f"Самый высокооплачиваемый: {top_employee['name']} ({top_employee['salary']})")
print(f"Отдел с лучшими зарплатами: {best_paid_dept}")
print(f"Сотрудников в 'high_salary.csv': {len(high_earners)}")
```

```python
import json


orders_data = [
    {"order_id": 1, "user": "Ali", "items": ["phone", "case"], "total": 300000},
    {"order_id": 2, "user": "Dana", "items": ["laptop"], "total": 800000},
    {"order_id": 3, "user": "Ali", "items": ["mouse", "keyboard"], "total": 70000}
]

with open('orders.json', 'w', encoding='utf-8') as f:
    json.dump(orders_data, f, indent=4)



with open('orders.json', 'r', encoding='utf-8') as f:
    orders = json.load(f)

total_revenue = 0
total_items_count = 0
user_order_counts = {}
item_frequency = {}
max_order_value = 0
top_user_expensive = ""

for order in orders:

    total_revenue += order['total']


    total_items_count += len(order['items'])


    if order['total'] > max_order_value:
        max_order_value = order['total']
        top_user_expensive = order['user']


    user = order['user']
    user_order_counts[user] = user_order_counts.get(user, 0) + 1


    for item in order['items']:
        item_frequency[item] = item_frequency.get(item, 0) + 1


most_popular_item = max(item_frequency, key=item_frequency.get)


summary = {
    "total_revenue": total_revenue,
    "top_user_by_expensive_order": top_user_expensive,
    "most_popular_item": most_popular_item,
    "total_orders": len(orders),
    "total_items_sold": total_items_count
}


with open('summary.json', 'w', encoding='utf-8') as f:
    json.dump(summary, f, ensure_ascii=False, indent=4)

print("Анализ завершен. Результаты в summary.json")
```

```python
import csv
import json


transactions_data = [
    ["user_id", "amount"],
    ["user_1", "5000"],
    ["user_2", "10000"],
    ["user_1", "700000"],
    ["user_3", "3000"],
    ["user_2", "900000"],
    ["user_4", "2000"],
    ["user_1", "1000"],
    ["user_1", "1500"]
]

with open('transactions.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(transactions_data)



suspicious_transactions_count = 0
suspicious_amount_total = 0
user_operations_count = {}
fraud_users = set()


with open('transactions.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        user_id = row['user_id']
        amount = int(row['amount'])


        if amount > 500000:
            suspicious_transactions_count += 1
            suspicious_amount_total += amount

            fraud_users.add(user_id)


        user_operations_count[user_id] = user_operations_count.get(user_id, 0) + 1


for user, count in user_operations_count.items():
    if count > 3:
        fraud_users.add(user)


with open('fraud_report.txt', 'w', encoding='utf-8') as report:
    report.write(f"Подозрительных транзакций (по сумме): {suspicious_transactions_count}\n")
    report.write(f"Подозрительных пользователей (всего): {len(fraud_users)}\n")
    report.write(f"Список пользователей: {', '.join(sorted(fraud_users))}\n")
    report.write(f"Общая сумма подозрительных операций: {suspicious_amount_total}\n")


with open('fraud_users.json', 'w', encoding='utf-8') as jf:

    json.dump({"fraud_users": list(fraud_users)}, jf, indent=4)

print("Антифрод-анализ завершен. Проверьте файлы fraud_report.txt и fraud_users.json.")
```
