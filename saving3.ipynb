```python
numbers = [1, 2, 3, 4, 5, 6]
nums=list(map(lambda x:x**2 if x%2==0 else x*3,numbers))
print(nums)
```

```python
def infinite_numbers(a):
    n=1
    while n<=a:
        if n%3==0 and n%5==0:
            yield "FizzBuzz"
        elif n%3==0:
            yield "Fizz"
        elif n%5==0:
            yield "Buzz"
        else:
            yield n
        n+=1
for x in infinite_numbers(15):
    print(x)
```

```python
numbers = [5, -2, 8, 0, -7, 3]
filtered=[x for x in numbers if x>=0]
res=list(map(lambda x: x/2 if x%2==0 else x*3+1,filtered))
for i in res:
    print(i)
```
