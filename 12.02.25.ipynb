```python
with open("data.txt","w", encoding="utf-8") as f:     #позволяет читать алфавит
    f.write("hello world")
with open("data.txt","r", encoding="utf-8") as f:
    print(f.read())
```

```python
with open("file.txt","w") as f:
    for i in range(1,11):
        f.write(str(i)+" ")
with open("file.txt","r") as f:
    print(f.read())
```

```python
names=["aset","Azamat","Eliya","Tattim"]
with open("names.txt","w") as f:
    for name in names:
        f.write(name + "\n")
with open("names.txt","r") as f:
    for line in f:
        print(line.strip().capitalize())
```

```python
foods=["Apple","banana","cherry"]
with open("foods.csv","w") as f:
    for food in foods:
        f.write(food + "\n")
with open("foods.csv","r") as f:
    print(f.read())
```
