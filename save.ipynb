```python
def filter_sets(sets_list):
    filtered_sets = []
    for sets in sets_list:
        found=False
        find=False
        for set in sets:
            if set>0:
                found=True
                break
            if found:
                if set%2==0:
                    filtered_sets.append(sets)
    return filtered_sets
nums=[{2, 4, 6, 8, 2}, {2, 4, 6}, {1, 3, 5, 7, 11}]
filtered_sets=filter_sets(nums)
print(filtered_sets)

```

```python
def update_counts(d,items):
    for item in items:
        if item in d:
            d[item]+=1
        else:
            d[item]=1
    return d
d = {}
items = ["apple", "banana", "apple", "cherry", "apple", "banana"]
update_counts(d,items)
```

```python
def union_of_filtered_sets(sets):
    filtered=set()
    for s in sets:
        for i in s:
            if i>10 and i%2!=0:
                if i not in filtered:
                    filtered.add(i)
    return filtered
b=[{11, 12, 13, 2}, {5, 15, 20, 25}, {11, 100, 101}]
union_of_filtered_sets(b)
```
