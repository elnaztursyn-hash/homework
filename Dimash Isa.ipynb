```python
import string

def check_password(password):
    result = {
        "length_ok": len(password) >= 10,
        "has_upper": any(c.isupper() for c in password),
        "has_lower": any(c.islower() for c in password),
        "has_digit": any(c.isdigit() for c in password),
        "has_special": any(c in string.punctuation for c in password),
        "no_spaces": " " not in password
    }

    # жалпы нәтиже
    result["is_strong"] = all(result.values())

    print(result)
```

```python
 students = [
    ("Айжан", 88),
    ("Марат", 95),
    ("Дана", 91),
    ("Ержан", 84),
    ("Алина", 97)
]

sorted_students = sorted(students, key=lambda x: x[1], reverse=True)

print("Сұрыпталған тізім:")
for s in sorted_students:
    print(s[0], "-", s[1])

top3 = sorted_students[:3]

print("\nҮздік 3 студент:")
for s in top3:
    print(s[0], "-", s[1])

average = sum(s[1] for s in top3) / len(top3)

print("\nҮздік 3 орташа бағасы:", average)
```

```python
sales = {
    "Almaty": {"Jan": 1200, "Feb": 1500},
    "Astana": {"Jan": 1000, "Feb": 1300}
}

city_totals = {city: sum(months.values()) for city, months in sales.items()}
print("Қалалар бойынша жалпы сатылым:")
for city, total in city_totals.items():
    print(city, "-", total)

month_totals = {}
for months in sales.values():
    for month, value in months.items():
        month_totals[month] = month_totals.get(month, 0) + value

print("\nАйлар бойынша жалпы сатылым:")
for month, total in month_totals.items():
    print(month, "-", total)

best_city = max(city_totals, key=city_totals.get)
print("\nЕң жақсы қала:", best_city)

best_month = max(month_totals, key=month_totals.get)
print("Ең жақсы ай:", best_month)
```

```python
user_A = {"python", "java", "docker", "linux"}
user_B = {"python", "aws", "linux", "kubernetes"}

common = user_A & user_B

only_A = user_A - user_B

only_B = user_B - user_A

print("Ортақ дағдылар:", common)
print("Тек A-де бар:", only_A)
print("Тек B-де бар:", only_B)
```

```python
user_perms = {"read", "write"}
required_perms = {"read", "write", "delete"}

missing_perms = required_perms - user_perms

extra_perms = user_perms - required_perms

has_all = missing_perms == set()

print("Барлық рұқсаттар жеткілікті:", has_all)
print("Жетіспейтін рұқсаттар:", missing_perms)
print("Артық рұқсаттар:", extra_perms)
```

```python
records = [
    {"id": 1, "email": "a@test.com", "tags": ["vip", "new"]},
    {"id": 2, "email": "b@test.com", "tags": ["vip"]},
    {"id": 3, "email": "a@test.com", "tags": ["fraud"]},
]

from collections import Counter

emails = [record["email"] for record in records]
email_counts = Counter(emails)
duplicate_emails = {email for email, count in email_counts.items() if count > 1}

print("Қайталанатын email-дер:", duplicate_emails)

from collections import defaultdict

email_tags = defaultdict(set)
for record in records:
    email_tags[record["email"]].update(record["tags"])
conflicting_tags = {email: tags for email, tags in email_tags.items() if len(tags) > 1}

print("Қақтығысқан тегтер:", conflicting_tags)
```
