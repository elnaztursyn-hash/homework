```python
#2.1
def even_numbers(n):
    for num in range(1,n+1):
        if num%2==0:
            if num%4==0:
                yield "kratno 4"
            else:
                yield num
for x in even_numbers(10):
    print(x)
```

```python
#2.2
def filter_words(words):
    for word in words:
        if len(word)>=4:
            if "a" in word.lower():
                yield "ca"
            else:
                yield word
words = ["cat", "car", "like", "model"]
for w in filter_words(words):
    print(w)
```

```python
#2.3
def infinite_numbers():
    num=1
    while True:
        res=""
        if num%3==0:
            res+="Fizz"
        if num%5==0:
            res+="Buzz"
        yield res if res else num
        num+=1
infinite_numbers(5)
```
